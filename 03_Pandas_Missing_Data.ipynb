```python
import numpy as np
import pandas as pd
```

**Finding Missing Data**

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

Any missing value in column. If yes than True

```python
df.isna().any()
```

**Removing Missing Data**

```python
df
```

Drop null rows

```python
df.dropna()
```

```python
df
```

`thresh=3` means atleast need 3 non-null values in a row

```python
df.dropna(thresh=3)
```

```python
df.dropna(thresh=1)
```

**Filling the missing Data**

```python
df
```

```python
df.fillna(0)
```

```python
values = {'A':0,'B':100,"C":300,'D':400}
df.fillna(value=values)
```

```python
df
```

```python
df.fillna(df.mean())
```
